```python
#2 No
#Separate the data points by class name
#df['Species']
setosa = df[df['Species'] == 'Iris-setosa']
setosa
```

```python
versicolor = df[df['Species'] == 'Iris-versicolor']
versicolor
```

```python
virginica = df[df['Species'] == 'Iris-virginica']
virginica
```

```python
#Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#Plot the data points for "setosa" using triangles
ax.scatter(setosa['SepalLengthCm'], setosa['SepalWidthCm'], setosa['PetalLengthCm'], marker='^', label='Setosa')

#Plot the data points for "Versicolor" using Square
ax.scatter(versicolor['SepalLengthCm'], versicolor['SepalWidthCm'], versicolor['PetalLengthCm'], marker='s', label='Versicolor')

#Plot the data points for "Virginica" using Circle
ax.scatter(virginica['SepalLengthCm'], virginica['SepalWidthCm'], virginica['PetalLengthCm'], marker='o', label='Virginica')

#Set Labels and titles for Scatter Plot
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.set_title('Iris Data Analysis')

#Add Legend
ax.legend()

#Show the 3D Plot
plt.show()
```

```python
#3 No
#Calculate the mean data point for each class
setosa_mean = df[df['Species'] == 'Iris-setosa'].mean()
versicolor_mean = df[df['Species'] == 'Iris-versicolor'].mean()
virginica_mean = df[df['Species'] == 'Iris-virginica'].mean()
```

```python
#Create a Scatter Diagram
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c='red', cmap='viridis', alpha=0.7)
```

```python
#Create a Scatter Diagram
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c='purple', cmap='viridis', alpha=0.7)
#Plot the mean data point with larger size
plt.scatter(setosa_mean['SepalLengthCm'], setosa_mean['SepalWidthCm'], marker='o', s=100, color='red', label="Setosa Mean")
plt.scatter(versicolor_mean['SepalLengthCm'], versicolor_mean['SepalWidthCm'], marker='o', s=100, color='green', label="Versicolor Mean")
plt.scatter(virginica_mean['SepalLengthCm'], virginica_mean['SepalWidthCm'], marker='o', s=100, color='blue', label="Virginica Mean")

#Set label and title
plt.xlabel('Sepal Length')
plt.ylabel("Sepal Width")
plt.title("Mean data points for each class")
plt.legend()
plt.show()
```

```python
#4 No
#Plot histogram for Sepal Length
plt.figure(figsize=(8, 6))
plt.hist(df['SepalLengthCm'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.grid(True)
plt.show()

#Plot histogram for Sepal Width
plt.figure(figsize=(8, 6))
plt.hist(df['SepalWidthCm'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.grid(True)
plt.show()
```

```python
#5 No
#Calculate the Standard deviation for 'Sepal Length' and 'Sepal Width'
std_sepal_length = np.std(df['SepalLengthCm'])
std_sepal_width = np.std(df['SepalWidthCm'])
print("Standard Deviation for Sepal Length is: ", std_sepal_length)
print("Standard Deviation for Sepal Width is: ", std_sepal_width)
```

```python
#6 No
sns.set(style='ticks')
#Create SubPlot for Sepal Length and Sepal Width
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

#Box plots for Sepal Length
sns.boxplot(data=df, x='Species', y='SepalLengthCm', ax=axes[0])
axes[0].set_title('Box Plot - Sepal Length')

#Box Plot for Sepal Width
sns.boxplot(data=df, x='Species', y='SepalWidthCm', ax=axes[1])
axes[1].set_title('Box Plot - Sepal Width')

#Adjust the layout
plt.tight_layout()

#Show the Plot
plt.show()
```

```python
#7NO
#Set the style of the plot
sns.set(style='ticks')

#Create subplots for Sepal Length and Sepal Width
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

#Plot box plots for Sepal Length for each class
sns.boxplot(x='Species', y='SepalLengthCm', data=df, ax=axes[0])
axes[0].set_title('Box Plot - Sepal Length')

#Box plot for sepal width for each class
sns.boxplot(x='Species', y='SepalWidthCm', data=df, ax=axes[1])
axes[1].set_title("Box Plot - Sepal Width")

#Adjust the layout
plt.tight_layout()

#Show the Plot
plt.show()
```

```python
#9 No
#Select the columns for the Parallel Coordinate Plot
parallel_data = df[['SepalLengthCm', 'SepalWidthCm', 'Species']]

#Create the parallel coordinate Plot
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(parallel_data, 'Species', color=['green', 'lightcoral', 'skyblue'])

#Add axis labels and ticks
plt.xlabel('Dimension')
plt.ylabel('Values')

#Add Legend
plt.legend()

#Show the Plot
plt.show()
```
